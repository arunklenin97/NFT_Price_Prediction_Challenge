# Final Solution 2

## Geometric Mean

```python
GM=S1.copy()
GM['last_sale_price']=(S1['last_sale_price']*S3['last_sale_price']*S5['last_sale_price'])**(1/3)
GM.head()
```

## Simple Mean 1

```python
SM1=S1.copy()
SM1['last_sale_price']=(S1['last_sale_price']
                        +S5['last_sale_price']
                        +S3['last_sale_price']
                        +GM['last_sale_price']+
                        f1['last_sale_price'])/5
SM1.head()
```

## Simple Mean2

```python
SM2=S1.copy()
SM2['last_sale_price']=(S1['last_sale_price']
                        +S2['last_sale_price']
                        +S4['last_sale_price']
                        )/3
SM2.head()
```

```python
f2=S1.copy()
f2['last_sale_price']=(S1['last_sale_price']
                        +S5['last_sale_price']
                        +S3['last_sale_price']
                        +f1['last_sale_price']
                        +GM['last_sale_price']
                        +SM1['last_sale_price']
                        +S5['last_sale_price']
                        +SM2['last_sale_price'])/8
f2.head()
```
